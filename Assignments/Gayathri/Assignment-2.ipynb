## Finding and replacing the outliers

### Outlier detection

```python
sns.boxplot(df.Age)
```

### Outlier removal-IQR method

```python
Q1 = df.Age.quantile(0.25)
Q3 = df.Age.quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
```

```python
upper_limit
```

```python
lower_limit
```

```python
df = df[df.Age<upper_limit]
```

```python
sns.boxplot(df.Age)
```

```python
df.shape
```

### Replacing the outliers using median

```python
sns.boxplot(df.Age)
```

```python
df.median()
```

```python
Q1 = df.Age.quantile(0.25)
Q3 = df.Age.quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
```

```python
df['Age'] = np.where(df['Age']>upper_limit,30,df['Age'])
```

```python
sns.boxplot(df.Age)
```

```python
df.shape
```

### Outlier removal-Percentile method

```python
sns.boxplot(df.Age)
```

```python
p99 = df.Age.quantile(0.96)
p99
```

```python
df = df[df.Age<=p99]
```

```python
sns.boxplot(df.Age)
```
